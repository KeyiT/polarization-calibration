```python
from scipy.io import loadmat
import os
import numpy as np
from sklearn import model_selection
from scipy.optimize import curve_fit
from sklearn import metrics
import matplotlib.pyplot as plt
```

# Load Data

```python
DATA_FOLER = 'data'
```

```python
x_json = loadmat(os.path.join(DATA_FOLER, 'example_calibrationfile.mat'))
```

```python
print(x_json)
```

```python
current = x_json['I']
topmw = x_json['TopmW']
```

```python
print(current.shape)
print(topmw.shape)
```

```python
current = np.squeeze(current)
topmw = np.squeeze(topmw)
```

```python
print(current.shape)
print(topmw.shape)
```

# Curve Fitting

## Split Training and Testing Data

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(current, topmw, test_size=0.25, random_state=3)
```

## Training 

```python
def func(x, a, b, c, d, alpha, beta):
    p = a * x**5 + b * x ** 4 + c * x**3 + d * x**2
    phi = alpha * p + beta
    return 0.5 * (1 + np.cos(phi))
```

```python
popt, pcov = curve_fit(func, x_train, y_train)
```

```python
popt
```

### Training r2

```python
metrics.r2_score(y_train,  func(x_train, *popt))
```

### Plot Training Data with Fitted Curve

```python
sort_indices = np.argsort(x_train)

plt.plot(x_train[sort_indices], y_train[sort_indices], 'b-', label='data')
plt.plot(x_train[sort_indices], func(x_train[sort_indices], *popt), 'g--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Testing

```python
y_pred = func(x_test, *popt)
```

```python
metrics.r2_score(y_test, y_pred)
```

### Plot Test Data with Fitted Curve

```python
sort_indices = np.argsort(x_test)

plt.plot(x_test[sort_indices], y_test[sort_indices], 'b-', label='data')
plt.plot(x_test[sort_indices], func(x_test[sort_indices], *popt), 'g--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
